```python
def RFECV_test(X,y):
    C = [10, 1, 0.1, 0.01, 0.001]
    step = [1, 5, 10, 15]
    cv = [2, 3, 4, 5]
    n=[]
    X_norm = sc.fit_transform(X.values)
    selected_features=[]
    for i in range(4):
        estimador = SVR(kernel='linear', C=C[i])
        for j in range(3):
            for k in range(3):
                selecionador = RFECV(estimador, step=step[j], cv=StratifiedKFold(cv[k]))
                selecionador = selecionador.fit(X_norm, y)
                X_features = pd.DataFrame(X.columns[selecionador.get_support()])
                selected_features.append(X_features.T.iloc[0])
                n.append(X_features.T.iloc[0].size)
    return n, selected_features
```
